## Models Integration with GOOGLE GEMINI and GROQ
This notebook demonstrates how to integrate multiple LLM providers with LangChain.

Topics covered:
- Groq model integration
- Google Gemini model integration
- invoke()
- stream()
- batch() with max_concurrency

```python
import os
from dotenv import load_dotenv

load_dotenv()
```

### Google Gemini model intergration

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("google_genai:gemini-2.5-flash")

response = model.invoke("Explain AI in simple words")
print(response.content)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
response= model.invoke("hi")
response
```

### GROQ MODEL integration

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:qwen/qwen3-32b")
response=model.invoke("tell me a short story")
response.content
```

```python
from langchain_groq import ChatGroq

model = ChatGroq(model="qwen/qwen3-32b")
response= model.invoke("tell me a joke")
response
```

## getting output using .stream()

```python
model.stream("tell me a joke in 100 words")
```

```python
for chunk in model.stream("why do parrots have colorful feathers"):
    print(chunk.text, end="")
```

## .batch()

```python
responses = model.batch([
    "how to airplane fly",
    "tell me about quantum computing ",
    "tell receipe to make cookie"
])

for response in responses:
    print(response)
```

```python
model.batch([
    "how to airplane fly",
    "tell me about quantum computing ",
    "tell receipe to make cookie"
],
config={
    'max_concurrency':5,
})
```
